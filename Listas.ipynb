```
numeros = list(map(float, input("Digite números separados por espaço: ").split()))

maior = max(numeros)
menor = min(numeros)
quantidade = len(numeros)

print("Maior número:", maior)
print("Menor número:", menor)
print("Quantidade de números digitados:", quantidade)
```
