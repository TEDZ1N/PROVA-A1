```
nota = float(input("Digite a nota do aluno (0 a 10): "))

if nota < 0 or nota > 10:
    print("Nota inválida! Digite um valor entre 0 e 10.")

elif nota < 5:
    print("Resultado: Reprovado")

elif nota < 7:
    print("Resultado: Recuperação")

else:
    print("Resultado: Aprovado")
```
